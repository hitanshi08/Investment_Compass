```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
company1_data = pd.read_csv("SBIN.csv")
company2_data = pd.read_csv("RELIANCE.csv")
```

```python
company1_total = company1_data['Close'].sum()
company2_total = company2_data['Close'].sum()

labels = ['Company 1', 'Company 2']
sizes = [company1_total, company2_total]
colors = ['#ff9999','#66b3ff']
explode = (0.1, 0) 

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  
plt.title('Total Closing Prices Comparison')
plt.show()
```

```python
total_volume_company1 = company1_data['Volume'].sum()
total_volume_company2 = company2_data['Volume'].sum()

labels = ['Company 1', 'Company 2']

sizes = [total_volume_company1, total_volume_company2]

colors = ['lightblue', 'lightgreen']

explode = (0, 0.1)  

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Volume Traded Comparison')
plt.axis('equal')  
plt.show()
```

```python
company1_stats = company1_data.describe()
company2_stats = company2_data.describe()
```

```python
print("Statistics for Company 1:")
print(company1_stats)
```

```python
print("\nStatistics for Company 2:")
print(company2_stats)
```

```python
company1_mean = company1_stats.loc['mean']
company2_mean = company2_stats.loc['mean']
company1_std = company1_stats.loc['std']
company2_std = company2_stats.loc['std']
```

```python
metrics = ['Open', 'High', 'Low', 'Close', 'Volume']
```

```python
fig, axs = plt.subplots(2, len(metrics), figsize=(18, 8), sharex=True)
```

```python
for i, metric in enumerate(metrics):
    axs[0, i].bar(['Company 1', 'Company 2'], [company1_mean[metric], company2_mean[metric]], color=['lightblue', 'lightgreen'])
    axs[0, i].set_title(f'{metric} - Mean')
    axs[0, i].set_ylabel('Value')
```

```python
for i, metric in enumerate(metrics):
    axs[1, i].bar(['Company 1', 'Company 2'], [company1_std[metric], company2_std[metric]], color=['lightblue', 'lightgreen'])
    axs[1, i].set_title(f'{metric} - Standard Deviation')
    axs[1, i].set_ylabel('Value')

plt.tight_layout()
plt.show()
```
